```python
##Quality
# year has many non-year values
# year is in objects --> Date time
# price has ask for price
# conver price into integer
# Km has nan values 
# kms has kms in every row 
# kms is in object --> into integer
# fuel type has some nan values
# keep first 3 words in name
```

```python
# CleanUp
backup = df.copy()
```

```python
df = df[df['year'].str.isnumeric()]
```

```python
df['year'] = df['year'].astype(int)
```

```python
df.info()
```

```python
df['Price'].unique()
```

```python
df = df[df['Price'] != 'Ask For Price']
```

```python
df['Price'].unique()
```

```python
df['Price'] = df['Price'].str.replace(',','').astype(int)
```

```python
df.info()
```

```python
df['kms_driven'] = df['kms_driven'].str.replace(',','')
df['kms_driven'] = df['kms_driven'].str.replace('kms','')
```

```python
df = df[df['kms_driven'] != 'Petrol']
```

```python
df['kms_driven'].unique()
```

```python
df['kms_driven'] = df['kms_driven'].str.replace(' ','')
df = df[df['kms_driven'] != '45000kms']
```

```python
df['kms_driven'] = df['kms_driven'].astype(int)
```

```python
df.info()
```

```python
df['fuel_type'].unique()
```

```python
df = df[~df['fuel_type'].isna()]
```

```python
df['fuel_type'].unique().sum()
```

```python
df['name'] = df['name'].str.split(' ').str.slice(0,3).str.join(' ')
```

```python
df = df.reset_index(drop=True)
```
